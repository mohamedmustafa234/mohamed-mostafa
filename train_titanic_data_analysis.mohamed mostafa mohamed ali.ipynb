# comments: 1- sex  is a strong feature as most of the survived people are females
# 2- pclass  is a strong feature as most of the survived passengers are from the first class 
# 3-  parch is a strong feature as well , as most of the survivers were the ones not having any children
# 4-  age  is a feature as well , as most of the survivers were adults 
# 5-  name , cabin , fare, ticket and ids are not features at all

```python
x=titanic[['Age','Pclass','Sex','Parch','SibSp']]
x['Age'].fillna(value=x['Age'].mean(),inplace=True)
x['Pclass'].fillna(value=x['Pclass'].mean(),inplace=True)
x['Parch'].fillna(value=x['Pclass'].mean(),inplace=True)
x['Sex'].fillna(value='male',inplace=True)
x['SibSp'].fillna(value=x['SibSp'].mean(),inplace=True)
x=x.iloc[:,:].values
```

```python
y=titanic.iloc[:,1]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
x[:, 2] = labelencoder.fit_transform(x[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [2])
x = onehotencoder.fit_transform(x).toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.neighbors import  KNeighborsClassifier
```

```python
regressor = KNeighborsClassifier(n_neighbors=7)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

a=confusion_matrix(y_test, y_pred)
a
```
